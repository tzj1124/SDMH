### An example showing the plot_elbow_curve method used by a scikit-learn clusterer

In this example, we'll be plotting an elbow curve for our k-means clusterer on the Iris dataset. We'll first create an instance of `KMeans` from scikit-learn, and then feed that to our `scikitplot.cluster.plot_elbow_curve` method.  
一个scikit-learn聚类使用plot_elbow_curve方法的例子

在本例中，我们将在Iris数据集上绘制k-means聚类的肘形曲线。我们将首先从scikit-learn创建一个“KMeans”实例，然后将其提供给我们的“scikitplot.cluster”。plot_elbow_curve”方法。

```python
from __future__ import absolute_import
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris as load_data

# Import scikit-plot
import scikitplot as skplt

%pylab inline
pylab.rcParams['figure.figsize'] = (14, 14)
```

```python
# Load the data
X, y = load_data(return_X_y=True)

# Create clusterer instance
kmeans = KMeans(random_state=1)
```

```python
# Plot!
skplt.cluster.plot_elbow_curve(kmeans, X, cluster_ranges=range(1, 11))
plt.show()
```
